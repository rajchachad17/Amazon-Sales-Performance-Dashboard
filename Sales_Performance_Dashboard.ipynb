### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Dataset

```python
df = pd.read_csv('data/raw/amazon (raw_data).csv')
df.head()
```

```python
df.info()
```

### Selecting Required Columns

```python
df = df[
     [
        "product_id",
        "product_name",
        "category",
        "discounted_price",
        "actual_price",
        "discount_percentage",
        "rating",
        "rating_count",
        "user_id",
        "review_id"
    ]
]
```

```python
df.columns.values
```

```python
#col_names = ['Product_id', 'Product_name', 'Category', 'Discounted_price', 'Actual_price', 'Discount_percentage', 'Rating', 'Rating_count', 'User_id', 'Review_id']
```

```python
df.head()
```

### Clean "Discounted Price", "Actual Price" Column

```python
df.columns.values
```

```python
df["discounted_price"] = (
    df["discounted_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df['discounted_price'][0:4]
```

```python
df["actual_price"] = (
    df["actual_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df['actual_price'][0:4]
```

```python
df["discount_percentage"] = (
    df["discount_percentage"]
    .str.replace("%", "", regex=False)
    .astype(float)
)
df["discount_percentage"][0:4]
```

### Classify Rating into categories & working with rating count

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df['rating_count'] = df['rating_count'].str.replace(',', '', regex=True)
df['rating_count']
```

### Handling Missing Values

```python
# Drop rows if core identity or pricing columns are missing
df = df.dropna(
    subset=[
        "product_id",
        "product_name",
        "discounted_price",
        "actual_price"
    ]
)
```

```python
df['category'] = df['category'].fillna('UNKNOWN_CATEGORY')

df['discount_percentage'] = (
    (df['actual_price'] - df['discounted_price']) / df['actual_price'] * 100
)

df['rating'] = df.groupby('category')['rating']\
                 .transform(lambda x: x.fillna(x.mean()))

df['rating_count'] = df['rating_count'].fillna(0)

df['user_id'] = df['user_id'].fillna('UNKNOWN_USER')

df['review_id'] = df['review_id'].fillna(0)
```

### Products per Category

```python
pd.read_sql(
    """
    SELECT category, COUNT(*) AS product_count
    FROM amazon_products
    GROUP BY category
    ORDER BY product_count DESC
    """,
    conn
)
```

### Average Rating by Category

```python
pd.read_sql(
    """
    SELECT category, ROUND(AVG(rating), 2) AS avg_rating
    FROM amazon_products
    GROUP BY category
    ORDER BY avg_rating DESC
    """,
    conn
)
```

### Top 10 highest-rated products

### Products with high discount but low rating

```python
pd.read_sql(
    """
    SELECT product_name, discounted_price, actual_price, discount_percentage
    FROM amazon_products
    WHERE discount_percentage > 40
    ORDER BY discount_percentage DESC
    LIMIT 10
    """,
    conn
)
```

### Best Price Segment

```python
pd.read_sql(
    """
    SELECT
        CASE
            WHEN discounted_price < 500 THEN 'Low'
            WHEN discounted_price BETWEEN 500 AND 2000 THEN 'Medium'
            ELSE 'High'
        END AS price_range,
        ROUND(AVG(rating), 2) AS avg_rating,
        COUNT(product_id) AS total_products
    FROM amazon_products
    GROUP BY price_range
    ORDER BY avg_rating DESC;
    """,
    conn
)
```

```python
pd.read_sql(
    """
    SELECT
        category,
        ROUND(AVG(rating), 2) AS avg_rating,
        COUNT(product_id) AS total_products
    FROM amazon_products
    GROUP BY category
    HAVING avg_rating >= 4.2
    ORDER BY total_products DESC;
    """,
    conn
)
```

### Rating Distribution

```python
pd.read_sql(
    """
    SELECT
        CASE
            WHEN rating >= 4.5 THEN 'Excellent'
            WHEN rating BETWEEN 4.0 AND 4.49 THEN 'Good'
            WHEN rating BETWEEN 3.0 AND 3.99 THEN 'Average'
            ELSE 'Poor'
        END AS rating_category,
        COUNT(product_id) AS product_count
    FROM amazon_products
    GROUP BY rating_category
    ORDER BY product_count DESC;
    """,
    conn
)
```

### Most Popular High-Rated Products

```python
pd.read_sql(
    """
    SELECT
        product_name,
        category,
        rating,
        rating_count,
        discount_percentage
    FROM amazon_products
    WHERE rating >= 4.5
    ORDER BY rating_count DESC
    LIMIT 5;
    """,
    conn
)
```

### Price Sensitivity Analysis

```python
pd.read_sql(
    """
    SELECT
        CASE
            WHEN discounted_price < 500 THEN 'Low'
            WHEN discounted_price BETWEEN 500 AND 2000 THEN 'Medium'
            ELSE 'High'
        END AS price_range,
        ROUND(AVG(discounted_price), 2) AS avg_price,
        ROUND(AVG(rating), 2) AS avg_rating
    FROM amazon_products
    GROUP BY price_range
    ORDER BY avg_price;
    """,
    conn
)
```
